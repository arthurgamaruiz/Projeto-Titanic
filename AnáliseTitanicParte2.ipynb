# Melhorando a acurácia da parte1


```python
# Importando pandas
import pandas as pd
```

```python
# Importando as base de treino
treino = pd.read_csv('/content/drive/MyDrive/Projeto Titanic/titanic/train.csv')
treino.head()
```

```python
# Importando a base de teste
teste = pd.read_csv('/content/drive/MyDrive/Projeto Titanic/titanic/test.csv')
teste.head()
```

```python
# Tipos de dados
treino.dtypes.value_counts()
```

Realizando o mesmo tratamento de dados que a parte 1


```python
# Eliminando colunas da base de treino
treino = treino.drop(['Name', 'Ticket', 'Cabin'], axis=1)    # axis=1 --> coluna
```

```python
# Eliminando colunas da base de teste
teste = teste.drop(['Name', 'Ticket', 'Cabin'], axis=1)
```

```python
# Localizar colunas com valores nulos e substituir pela média
treino.loc[treino.Age.isnull(), 'Age'] = treino.Age.mean()
```

```python
# Repetindo para a base de teste
teste.loc[teste.Age.isnull(), 'Age'] = teste.Age.mean()
```

```python
# Substituindo os valores vazios
treino.loc[treino.Embarked.isnull(), 'Embarked'] = treino.Embarked.mode()[0]

# Base de teste não possui valores nulos na coluna Embarked
```

```python
teste.loc[teste.Fare.isnull(), 'Fare'] = teste.Fare.mean()
```

Entendendo as colunas de texto


```python
# Verificando as colunas de texto
treino.columns[treino.dtypes == 'object']
```

```python
# Valores na coluna Sex
treino.Sex.value_counts()
```

```python
# Valores na coluna Embarked
treino.Embarked.value_counts()
```

Tratando a coluna Sex, criando uma nova coluna (MaleCheck) que será 1 para homem e 0 para mulher

```python
# Lambda function para fazer o tratamento
treino['MaleCheck'] = treino.Sex.apply(lambda x: 1 if x == 'male' else 0)
treino.head()
```

```python
treino[['Sex', 'MaleCheck']].value_counts()
```

```python
# Fazer o mesmo para a base de teste
teste['MaleCheck'] = teste.Sex.apply(lambda x: 1 if x=='male' else 0)
teste
```

```python
# Verificando valores
teste[['Sex', 'MaleCheck']].value_counts()
```

Agora tratar a coluna Embarked utilzando [OneHotEncoder](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html) que criará uma nova coluna para cada um dor rótulos da coluna Embarked

```python
# Importando o OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
# Criando o encoder
enc = OneHotEncoder(handle_unknown='ignore')
```

```python
# Fazendo o ajuste com os dados
enc.fit(treino[['Embarked']])
```

```python
enc.categories_
```

```python
# Fazendo a transformação
enc.transform(treino[['Embarked']]).toarray()
```

```python
# Transformando esse resultado em um DataFrame
enc_df = pd.DataFrame(enc.transform(treino[['Embarked']]).toarray(), columns = enc.get_feature_names_out())
# columns --> nome das colunas
enc_df.head()
```

```python
# Concatenando ambos os df
treino = pd.concat([treino, enc_df], axis=1)
treino
```

```python
# Verificando valores
treino[['Embarked','Embarked_C', 'Embarked_Q', 'Embarked_S']].value_counts()
```

Fazendo o mesmo para a base de teste

```python
teste.head(3)
```

```python
# Transformando esse resultado em um DataFrame
enc_df = pd.DataFrame(enc.transform(teste[['Embarked']]).toarray(), columns = enc.get_feature_names_out())
# columns --> nome das colunas
enc_df.head()
```

```python
# Concatenando ambos os df
teste = pd.concat([teste, enc_df], axis=1)
teste
```

```python
# Verificando valores
teste[['Embarked','Embarked_C', 'Embarked_Q', 'Embarked_S']].value_counts()
```

Utilizando essa nova base para aplicar no modelo

```python
# Visualizando a base
treino.head(3)
```

```python
# Apagando as colunas tratadas
treino = treino.drop(['Sex', 'Embarked'], axis=1)
treino
```

```python
# Apagando as colunas tratadas
teste = teste.drop(['Sex', 'Embarked'], axis=1)
teste.head(3)
```

Aplicando os modelos

```python
# Importanto train_test_split
from sklearn.model_selection import train_test_split
```

```python
# Separando a base de treino em X e y
X = treino.drop(['PassengerId', "Survived"], axis=1)
y = treino.Survived
```

```python
# Separando em treino e validação
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.33, random_state=42)
```

Árvore de classificação

```python
# Fazendo a importação
from sklearn import tree
```

```python
# Criando o Classifier
clf_ac = tree.DecisionTreeClassifier(random_state=42)
```

```python
# Fazendo o ajuste com os dados
clf_ac.fit(X_train, y_train)
```

```python
# Fazendo a previsão
y_pred_ac = clf_ac.predict(X_val)
```

KNeighboorsClassifier

```python
# Importação
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Criando o classificador
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Fazendo o ajuste com os dados
knn.fit(X_train, y_train)
```

```python
# Fazendo a previsão
y_pred_knn = knn.predict(X_val)
```

Regressão logística

```python
# Importando
from sklearn.linear_model import LogisticRegression
```

```python
# Criando classificador
clf_rl = LogisticRegression(random_state=42)
```

```python
# Fazendo o ajuste com os dados
clf_rl.fit(X_train, y_train)
```

```python
# Fazendo a previsão
y_pred_rl = clf_rl.predict(X_val)
```

Avaliando os modelos (acurácia)

```python
# Importando
from sklearn.metrics import accuracy_score
```

```python
# Para a árvore
accuracy_score(y_val, y_pred_ac)
```

```python
# Para o knn
accuracy_score(y_val, y_pred_knn)
```

```python
# Para a regressão logística
accuracy_score(y_val, y_pred_rl)
```

Matriz de confusão

```python
# Importando
from sklearn.metrics import confusion_matrix
```

```python
# Para a árvore
confusion_matrix(y_val, y_pred_ac)
```

```python
# Para o knn
confusion_matrix(y_val, y_pred_knn)
```

```python
# Para a regressão logística
confusion_matrix(y_val, y_pred_rl)
```
